```python
import xml.etree.ElementTree as ET

path = "dir where mod_tables.xml is"  # or mod_tables.xml
tree = ET.parse(path)
root = tree.getroot()

def local(tag): return tag.split('}', 1)[-1] if '}' in tag else tag

rows = [el for el in root.iter() if local(el.tag) == "modifications_row"]
print("rows:", len(rows))
for i, row in enumerate(rows[:5]):
    print(f"\n--- modifications_row #{i} ---")
    for k, v in row.attrib.items():
        print(f"{k} = {v}")
```
